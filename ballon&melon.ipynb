```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
N = 500
std = 0.6

np.random.seed(121)

melons = pd.DataFrame({'height': np.random.normal(8,std,N),
                       'width': np.random.normal(5,std,N),
                       'weight': np.random.normal(10,std,N)})

ballons = pd.DataFrame({'height': np.random.normal(9,std,N),
                       'width': np.random.normal(7,std,N),
                       'weight': np.random.normal(1,std,N)})

fruits = pd.concat([melons,ballons])
```

```python
figure = plt.figure()
threeD_figure = figure.add_subplot(projection='3d')

threeD_figure.scatter3D(melons['height'], melons['width'], melons['weight'], color= 'green')
threeD_figure.scatter3D(ballons['height'], ballons['width'], ballons['weight'], color= 'red')

threeD_figure.set_xlabel('height')
threeD_figure.set_ylabel('width')
threeD_figure.set_zlabel('weight')

plt.show()
```
